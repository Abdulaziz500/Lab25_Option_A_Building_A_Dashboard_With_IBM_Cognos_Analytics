```python
df = df[df['Age'] != 'Prefer not to say']
df['Age'].unique()
```

```python
# Remove irrelevant values (e.g., "Nomadic")
irrelevant_countries = ['Nomadic']
df = df[~df['Country'].isin(irrelevant_countries)]

# Reset index after cleaning
df.reset_index(drop=True, inplace=True)

# Define mapping for long country names
country_fixes = {
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'Iran, Islamic Republic of...': 'Iran',
    'Venezuela, Bolivarian Republic of...': 'Venezuela',
    "Democratic People's Republic of Korea": 'North Korea',
    'Republic of Korea': 'South Korea',
    'Congo, Republic of the...': 'Republic of the Congo',
    'Hong Kong (S.A.R.)': 'Hong Kong',
    'Micronesia, Federated States of...': 'Micronesia',
    'Libyan Arab Jamahiriya': 'Libya',
    "Lao People's Democratic Republic": 'Laos',
}

# Apply replacements
df['Country'] = df['Country'].replace(country_fixes)
df['Country']
```

```python
# # # Define mapping for edlevel column
edlevel_fixes = {
    'Primary/elementary school': 'Primary',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary',
    'Some college/university study without earning a degree': 'Some College',
    'Associate degree (A.A., A.S., etc.)': 'Associate',
    "Bachelor's degree (B.A., B.S., B.Eng., etc.)": "Bachelor’s",
    "Master's degree (M.A., M.S., M.Eng., MBA, etc.)": "Master’s",
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Doctorate',
}

# # Map the EdLevel column
df['EdLevel'] = df['EdLevel'].replace(edlevel_fixes)
df['EdLevel']
```

```python
df.to_csv("survey_data_updated_4.csv", index=False)
```
